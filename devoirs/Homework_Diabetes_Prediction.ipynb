- **BMI:** Indice de masse corporelle (poids en kg/(taille en m)²)
- **DiabetesPedigreeFunction:** Fonction de prédisposition génétique au diabète
- **Age:** Âge (années)
- **Outcome:** Variable cible (0 = pas de diabète, 1 = diabète)
